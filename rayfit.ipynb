```python
# Import required packages
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
```

```python
def exponential(x, a, b,c):
    return a*np.exp(-b*x)+c
```

```python
import pandas as pd
df = pd.read_excel("prueba.xlsx")
df['Medida']
```

```python
pars, cov = curve_fit(f=exponential, xdata=df['Medida'], ydata=df['T'], p0=[0, 0,0], bounds=(-np.inf, np.inf))
# Get the standard deviations of the parameters (square roots of the diagonal of the covariance)
stdevs = np.sqrt(np.diag(cov))

# Calculate the residuals
res = df['T'] - exponential(df['Medida'], *pars)
```

```python
# Create figure object and store it in a variable called 'fig'
fig = plt.figure(figsize=(3, 3))
# Add axes object to our figure that takes up entire figure
ax = fig.add_axes([0, 0, 1, 1])
ax.scatter(df['Medida'],df['T'], color='#00b3b3', label='Data')
ax.plot(df['Medida'], exponential(df['Medida'], *pars), linestyle='--', linewidth=2, color='black')
pars
```
